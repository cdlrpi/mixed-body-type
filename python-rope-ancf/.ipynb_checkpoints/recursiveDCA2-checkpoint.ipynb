```python
# Recursive DCA
#---------------
# This is a script to solve for the motion of a 
# massive rod double pendulum using DCA

import numpy as np
import math
from scipy.integrate import odeint
import matplotlib.pyplot as plt

import MBstructs as MB
import MultiBodyFuncts as MBF
from dca_recursive import dca_recursive

%matplotlib inline
```

```python
# System Definition
#--------------------

#number of bodies
n=7

#Create a list of Bodies and Joints
bodies=[]
joints=[]

for i in range(0,n):
    bodies.append(MB.Rigid_Body())
    joints.append(MB.Joint(np.array((0,0,1,0,0,0)),1))

#Body Properties
for i in range (0,n):
    bodies[i].m=1.0
    bodies[i].l=1.0
    bodies[i].Inertia(bodies[i].m*(bodies[i].l**2)/12,1,bodies[i].m*(bodies[i].l**2)/12)

```

```python
# Initial Conditions
#--------------------
# q1 and q2 are measured as the angle the rod makes with -n2 axis
q_init=np.zeros((n))
q_init[0]=np.pi/2

qdot_init=np.zeros((n))

initialconditions=np.zeros((2*n))
initialconditions[:n]=q_init
initialconditions[n:]=qdot_init

#Length of time of the simulation

Time=np.arange(0,5,.01)
```

```python
# Initialization Function
#-------------------------
#This function prepares the system for the DCA algorithm 

def initialize(bodies,x,n):
    val1=0
    val2=np.zeros((3))
    val3=np.zeros((3))
    for k in range(0,n):
        val1=val1+x[k]
        bodies[k].C0=MBF.simdcm(val1,np.array((0,0,1)))
        val2=val2+np.array((0,0,x[k+n]))
        bodies[k].w=val2
        bodies[k].w1=np.array((0,0,x[k+n]))
        bodies[k].rs(bodies[k].l/2.0,np.array((0,-1,0)))
        bodies[k].Vcm=val3+np.cross(bodies[k].w,bodies[k].r10)
        val3=val3+np.cross(bodies[k].w,bodies[k].r12)
        bodies[k].Inertia2()#transformed inertia matrix
        bodies[k].Mmatrix()#Mass matrix and its inverse
        bodies[k].Forces(np.array((0,-1,0)))
        bodies[k].shifters()#S matrices
        bodies[k].zeta()#zeta values
```

```python
# Helper Function
#-------------------------
#This function provides a buffer between the dca algorithm and the 
#integrator call. 
def funct(state,t,n,i,bodies,joints,BC1,BC2):

    #Initialize the system
    initialize(bodies,state,n)
    
    #Call the Recursive DCA Algorithm
    #This returns a list of the form:
    #[A11,A12,A21,A22,...,An1,An2]
    #where Axy corresponds to the acceleration
    #of the yth handle of the xth body
    ll=dca_recursive(n,0,bodies,joints,BC1,BC2,state)
    #loop to fill d_dt with the acceleration values

    d_dt=np.zeros((2*n))
    for j in range(0,n):
        if j == 0:
            A1=ll.pop(0)
            d_dt[j+n]=np.dot(np.transpose(joints[j].P),A1)
        else:
            A2= ll.pop(0)
            A1=ll.pop(0)
            d_dt[j+n]=np.dot(np.transpose(joints[j].P),(A1-A2))
    
    #add the velocities to d_dt and return to the integrator
    d_dt[:n]=state[n:]
    return d_dt 
```

```python
# Integration
#-------------

#odeint is the numerical integrator used
yy = odeint(funct,initialconditions,Time,(n,0,bodies,joints,2,1))
# print(yy.astype(np.float16))
```

```python
# Energy Calculation
#--------------------
#The energy of the system is calculated and plotted

energy=MBF.PendEnergy(yy,bs)
KE=energy[:,0]
PE=energy[:,1]
TE=energy[:,2]

plt.plot(Time,TE-TE[0])
plt.xlabel("Time [s]")
plt.ylabel("energy")
plt.title("System Energy")
plt.show()

plt.plot(Time,PE,Time,KE)
plt.xlabel("Time[s]")
plt.ylabel("energy")
plt.title("Kinetic and Potential Energy")
plt.show
```

```python
# Solution Plot
#--------------------
plt.plot(Time,yy[:,:n])
plt.xlabel("Time [s]")
plt.ylabel("Generalized Coordinates [Rad]")
plt.title("System Response")

plt.show()

plt.plot(Time,yy[:,n:])
plt.xlabel(("Time[s]"))
plt.ylabel(("Generalized Speeds [Rad/s]"))
plt.title("System Response")

plt.show()
```
